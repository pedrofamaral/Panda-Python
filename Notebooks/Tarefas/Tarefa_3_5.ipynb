#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```
import pandas as pd
df = pd.read_csv('/content/DSBD/Datasets/Hourly_wages/hourly_wages.csv')
df.head()
df.tail()
```

#### c-) Mostrar os histogramas dos atributos: *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].hist(df['education_yrs'], bins=10)
axes[0].set_title('Histograma de education_yrs')
axes[0].set_xlabel('education_yrs')
axes[0].set_ylabel('Frequência')
axes[1].hist(df['experience_yrs'], bins=10)
axes[1].set_title('Histograma de experience_yrs')
axes[1].set_xlabel('experience_yrs')
axes[1].set_ylabel('Frequência')
axes[2].hist(df['age'], bins=10)
axes[2].set_title('Histograma de age')
axes[2].set_xlabel('age')
axes[2].set_ylabel('Frequência')
plt.tight_layout()
plt.show()
```

#### d-) Mostrar os gráficos de setores (*pie chart*) de cada um dos atributos: *union*, *female*, *marr*, *south*, *manufacturing* e *construction* num mesmo quadro de plotagem (usando subplots).

```
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].pie(df['union'].value_counts(), labels=['Não', 'Sim'], autopct='%1.1f%%')
axes[0, 0].set_title('Filiação Sindical')

axes[0, 1].pie(df['female'].value_counts(), labels=['Masculino', 'Feminino'], autopct='%1.1f%%')
axes[0, 1].set_title('Sexo')

axes[0, 2].pie(df['marr'].value_counts(), labels=['Solteiro', 'Casado'], autopct='%1.1f%%')
axes[0, 2].set_title('Estado Civil')

axes[1, 0].pie(df['south'].value_counts(), labels=['Norte/Outros', 'Sul'], autopct='%1.1f%%')
axes[1, 0].set_title('Região')

axes[1, 1].pie(df['manufacturing'].value_counts(), labels=['Não', 'Sim'], autopct='%1.1f%%')
axes[1, 1].set_title('Indústria')

axes[1, 2].pie(df['construction'].value_counts(), labels=['Não', 'Sim'], autopct='%1.1f%%')
axes[1, 2].set_title('Construção')

plt.tight_layout()
plt.show()
```

#### e-) Mostrar os boxplots das variáveis *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].boxplot(df['education_yrs'])
axes[0].set_title('Boxplot de education_yrs')
axes[0].set_ylabel('education_yrs')
axes[1].boxplot(df['experience_yrs'])
axes[1].set_title('Boxplot de experience_yrs')
axes[1].set_ylabel('experience_yrs')
axes[2].boxplot(df['age'])
axes[2].set_title('Boxplot de age')
axes[2].set_ylabel('age')
plt.tight_layout()
plt.show()
```

#### f-) Mostre lado a lado os boxplots da variável *age* discriminados por sexo (female).

```
plt.figure(figsize=(8, 6))
sns.boxplot(x='female', y='age', data=df)
plt.xlabel('Sexo (0: Masculino, 1: Feminino)')
plt.ylabel('Idade')
plt.title('Boxplot da Idade por Sexo')
plt.show()
```

#### h-) Mostre lado a lado os boxplots da variável *wage_per_hour* discriminados por sexo (female).

```
plt.figure(figsize=(8, 6))
sns.boxplot(x='female', y='wage_per_hour', data=df)
plt.xlabel('Sexo (0: Masculino, 1: Feminino)')
plt.ylabel('Salário por Hora')
plt.title('Boxplot do Salário por Hora por Sexo')
plt.show()
```

Resposta:O salário médio dos homens é superior ao das mulheres, com maior variação entre os valores no grupo masculino. No entanto, no grupo feminino, há um número maior de pessoas com salários que fogem ao padrão.








#### j-) Mostre lado a lado os boxplots da variável wage_per_hour discriminados por estado civil (marr).

```
plt.figure(figsize=(8, 6))
sns.boxplot(x='marr', y='wage_per_hour', data=df)
plt.xlabel('Estado Civil (0: Solteiro, 1: Casado)')
plt.ylabel('Salário por Hora')
plt.title('Boxplot do Salário por Hora por Estado Civil')
plt.show()
```
